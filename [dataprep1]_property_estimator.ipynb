##เปรียบเทียบความแตกต่างในการประเมินราคาทรัพย์แต่ละทรัพย์ของนักประเมินแต่ละคนและของราคาตั้งขาย
#ใช้ MAPE เป็นเกณฑ์ในการเปรียบเทียบ
#hint เริ่มต้นมาเราอยากได้ตารางรูปแบบนี้ก่อน

###บ้าน

```python
est2['ประเมินโดย1'] = est2['post id'].map(est1.set_index('post id')['ประเมินราคาซื้อ-ขาย']);
est2
```

```python
est2.rename(columns={'ประเมินราคาซื้อ-ขาย':'ประเมินโดย2'},inplace=True);
est2
```

```python
home1 = est2[['post id','ราคาทรัพย์','ประเมินโดย1','ประเมินโดย2']]
home1
```

```python
ans_h = est2[['post id','ราคาทรัพย์','ประเมินโดย1','ประเมินโดย2']]
ans_h
```

```python
ans_h.isnull().any() # เช็คค่าว่าง
```

```python
ans_h1 = ans_h.dropna(subset=['ประเมินโดย2']) # ลบค่าว่างออก
ans_h1
```

```python
ans_h1.isnull().any() #เช็คค่าว่างอีกรอบ
```

###MAPE บ้าน

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mape_h1 = mean_absolute_percentage_error(ans_h1['ราคาทรัพย์'],ans_h1['ประเมินโดย1'])
print(mape_h1)
```

```python
mape_h2 = mean_absolute_percentage_error(ans_h1['ราคาทรัพย์'],ans_h1['ประเมินโดย2'])
print(mape_h2)
```
